```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
```

```python
train_set= pd.read_csv('/content/wheat_train.csv')
train_set_x= train_set.drop(['ID','Type'],axis=1)
train_set_x.head()
```

```python
train_set_y=train_set['Type']
train_set_y.head()
```

```python
test_set=pd.read_csv('/content/wheat_test.csv')
tst= test_set.drop(['ID'],axis=1)
tst.head()
```

```python
from sklearn import linear_model
from sklearn.linear_model import LogisticRegression
regress=linear_model.LogisticRegression()
regress.fit(train_set_x,train_set_y)
print ('Coefficients: ', regress.coef_)
print ('Intercept: ',regress.intercept_)
```

```python
type_predicted=regress.predict(train_set_x)
print(type_predicted)
```
