```python
import pandas as pd
import numpy as np
```

```python
import sklearn as sk
from matplotlib import pyplot as plt
```

```python
df=pd.read_csv("/content/stu (1).csv")
df.head() #td is training data
```

```python
Y_train = df.cgpa[0:25]
df1 = df.drop(['name','cgpa'],axis = 1)
X_train = df1[0:25]
X_test = df1[25:40]
X_train
```

```python
Y_train
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg = reg.fit(X_train,Y_train)
```

```python
Y_test = reg.predict(X_test)
```

```python
Y_test
```

```python
r2_scores = reg.score(X_train,Y_train)
```

```python
r2_scores
```

```python
plt.scatter(X_train,Y_train, color = 'brown')
plt.plot(X_test,Y_test,color="brown")
```
