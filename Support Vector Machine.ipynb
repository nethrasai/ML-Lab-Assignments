```python
import numpy as np # linear algebra
import pandas as pd
import matplotlib.pyplot as plt
```

```python
train_data = pd.read_csv("/content/train (1).csv")
train_data.head()
```

```python
train_data.dtypes
```

```python
train_data['Location'] = train_data['Location'].astype('category')
train_data['Urban'] = train_data['Urban'].astype('category')
train_data['US'] = train_data['US'].astype('category')

train_data["Loc_category"]=train_data['Location'].cat.codes
train_data['Urban_category']=train_data['Urban'].cat.codes
train_data['US_category']=train_data['US'].cat.codes
train_data.head()
```

```python
category = pd.cut(train_data.Sales,bins=[0,5,10,50],labels=['1','2','3'])
train_data.insert(1, 'Class', category)
```

```python
train_drop = train_data.drop(['Sales','Location','Urban','US'], axis=1)
train_drop.head()
```

```python
nan_dropped_data=train_drop.dropna()
```

```python
X = nan_dropped_data.drop(['Class'],axis=1)
y = nan_dropped_data['Class']
```

```python
from sklearn.svm import SVC
```

```python
svclassifier = SVC(kernel='linear')
svclassifier.fit(X,y)
```

```python
test_data = pd.read_csv('/content/test (1).csv')
test_data.head()
```

```python
test_data['Location'] = test_data['Location'].astype('category')
test_data['Urban'] = test_data['Urban'].astype('category')
test_data['US'] = test_data['US'].astype('category')

test_data["Loc_category"]=test_data['Location'].cat.codes
test_data['Urban_category']=test_data['Urban'].cat.codes
test_data['US_category']=test_data['US'].cat.codes
test_data.head()
```

```python
category_test_data = pd.cut(test_data.Sales,bins=[0,5,10,50],labels=['1','2','3'])
test_data.insert(1, 'Class', category_test_data)
```

```python
test_drop = test_data.drop(['Sales','Location','Urban','US'], axis=1)
test_drop.head()
```

```python
tst=test_drop.drop(['Class'],axis=1)
print(tst)
```

```python
y_pred = svclassifier.predict(tst)
print(y_pred)
```

```python
id=test_data['ID']
```

```python
import csv
submission = open("20BRS1092_SVM.csv","w")
sub_file = csv.writer(submission)
sub_file.writerow(['ID','Class'])
for i in range(0,len(y_pred)):
    sub_file.writerow([str(id[i]),str(y_pred[i])])

submission.close()
```
